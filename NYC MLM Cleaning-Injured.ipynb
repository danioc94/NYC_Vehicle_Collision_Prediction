```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(Path('CSV/Motor_Vehicle_Collisions_-_Crashes.csv'))
df.head()
```

```python
print(df.columns)
```

```python
df.rename(columns={'CRASH TIME': 'CRASHTIME'})
```

```python
injured_df =df.drop(columns=['LATITUDE',
       'LONGITUDE', 'LOCATION', 'ON STREET NAME', 'CROSS STREET NAME',
       'OFF STREET NAME',
       'NUMBER OF PERSONS KILLED', 'NUMBER OF PEDESTRIANS INJURED',
       'NUMBER OF PEDESTRIANS KILLED', 'NUMBER OF CYCLIST INJURED',
       'NUMBER OF CYCLIST KILLED', 'NUMBER OF MOTORIST INJURED',
       'NUMBER OF MOTORIST KILLED', 'CONTRIBUTING FACTOR VEHICLE 1',
       'CONTRIBUTING FACTOR VEHICLE 2', 'CONTRIBUTING FACTOR VEHICLE 3',
       'CONTRIBUTING FACTOR VEHICLE 4', 'CONTRIBUTING FACTOR VEHICLE 5',
       'COLLISION_ID', 'VEHICLE TYPE CODE 1', 'VEHICLE TYPE CODE 2',
       'VEHICLE TYPE CODE 3', 'VEHICLE TYPE CODE 4', 'VEHICLE TYPE CODE 5'], axis=1)
injured_df
```

```python
injured_df.dropna(axis=0,how='any',inplace=True)
```

```python
injured_df.dtypes
```

```python
injured_df['CRASH DATE']= pd.to_datetime(df['CRASH DATE'])
```

```python
injured_df.dtypes
```

```python
year2017_df = injured_df.loc[injured_df['CRASH DATE'].dt.year == 2017]
```

```python
year2017_df
```

```python
output_file_path = "CSV/year2017.csv"
year2017_df.to_csv(output_file_path, index=False)
```
